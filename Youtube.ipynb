Python Data Cleaning and Visualization Report:

The dataset has been cleaned across all parameters. For the 'video_views' column, I applied the clip method to trim extreme values and used an outlier removal technique to ensure cleaner data. All other numerical columns were retained in their original form to preserve the integrity of the dataset, as the goal was to clean the data for dashboard creation rather than deleting records.

For visualization, I created Power BI dashboards that allow effective data comparison and provide insightful, visually appealing representations of the dataset.

```python
df.drop(['created_month','created_date'],inplace=True, axis=1)
```

“I removed the subscribers_for_last_30_days column because around one-third of the data was missing, which could introduce bias. Since other stable subscriber metrics were available, dropping the column improved data quality.”

```python
df.drop(['subscribers_for_last_30_days'],inplace=True, axis=1)
```

```python
df['created_year'] = df['created_year'].astype('Int64')
print(df.dtypes)
```

```python
len(df.columns)
```

```python
df['category'].fillna('Unknown', inplace=True)
df['Country'].fillna('Unknown', inplace=True)
df['channel_type'].fillna('Unknown', inplace=True)
```

```python
df['video_views_rank'].fillna(df['video_views_rank'].median,inplace=True)
```

```python
import pandas as pd

df['video_views_rank'] = pd.to_numeric(
    df['video_views_rank'],
    errors='coerce'
)
```

```python
df['video_views_rank'].fillna(
    df['video_views_rank'].median(),
    inplace=True
)
```

```python
df['video_views_for_the_last_30_days'] = (
    df['video_views_for_the_last_30_days']
    .fillna(df['video_views_for_the_last_30_days'].median())
)
```

```python
df['created_year'] = (
    df['created_year']
    .fillna(df['created_year'].mode()[0])
)
```

```python
df['Urban_population']= df['Urban_population'].fillna(df['Urban_population'].median())
```

```python
df['Population']= df['Population'].fillna(df['Population'].median())
```

All missing values are filled category column with UNKNOWN and numerical data with MEDIAN/MODE

```python
def calculation(df):

    cols = [
        'video views',
        'video_views_for_the_last_30_days',
        'lowest_monthly_earnings',
        'highest_monthly_earnings',
        'lowest_yearly_earnings',
        'highest_yearly_earnings',
        'Population',
        'Urban_population'
    ]

    col = df[cols]

    summary = col.agg(['mean', 'median', 'var', 'std', 'skew']).T
    summary['CV'] = summary['std'] / summary['mean']

    print(summary.round(2))


calculation(df)
```

```python
Q1_video_views= df['video views'].quantile(0.25)
Q3_video_views=df['video views'].quantile(0.75)
IQR=Q3_video_views-Q1_video_views
print(IQR)
```

```python
Lower_fence= Q1_video_views-1.5*IQR
print(Lower_fence)
Upper_fence=Q3_video_views+1.5*IQR
print(Upper_fence)
```

```python
outliers = df[(df['video views'] < Lower_fence) | (df['video views'] > Upper_fence)]
outliers
##outlier = df[((cols < Lower_fence) | (cols > Upper_fence)).any(axis=1)]
```

```python
print("Total outliers:", len(outliers))
```

```python
Q1= df['video_views_for_the_last_30_days'].quantile(0.25)
Q3=df['video_views_for_the_last_30_days'].quantile(0.75)
IQR= Q3-Q1
print(IQR)
```

```python
Lf=Q1-1.5*IQR
Uf=Q3+1.5*IQR
print(Lf,Uf)
```

```python
outliers = df[(df['video_views_for_the_last_30_days'] < Lf) | (df['video_views_for_the_last_30_days'] > Uf)]
```
